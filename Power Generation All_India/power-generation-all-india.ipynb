# Power Generation data All India

### Based on the months(Feb-2019 to May-2023)

| Column Name               | Description                                     |
|---------------------------|-------------------------------------------------|
| Month-Year                | Month and Year of the power generation data      |
| Thermal                   | Power generated from thermal sources             |
| Nuclear                   | Power generated from nuclear sources             |
| Hydro                     | Power generated from hydro sources               |
| Renewable                 | Power generated from renewable sources           |
| Total Power Generation    | Total power generation in Billion Units (BU)     |


### Purpose 
* Important Trends Analysed Based on EDA
* Exploring th Reasons Of the trends 
*  Giving Recommendation for future Work

```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as  plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
df=pd.read_csv('/kaggle/input/all-india-power-generation-data/All India power generation.csv')
```

```python
df=df.fillna(df.mean())
```

```python
df.describe()
```

```python
#poweer generation over time for different diffrent sources
plt.plot(df['Month-Year'], df['Nuclear'])
plt.xlabel('Month-Year')
plt.ylabel('Power Generation (BU)')
plt.title('Power Generation from Nuclear sources over time')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df['Month-Year'],df['Thermal'])
plt.xlabel('Date')
plt.ylabel('Power Generation(BU)')
plt.title('Power Generation from Thermal sources over time')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.plot(df['Month-Year'], df['Hydro'])
plt.xlabel('Month-Year')
plt.ylabel('Power Generation (BU)')
plt.title('Power Generation from Hydro sources over time')
plt.xticks(rotation=45)
plt.show()
```

```python
#correlation between power generation from Hydro and Renewable sources
sns.scatterplot(data=df, x='Hydro', y='Renewable')
plt.xlabel('Power Generation from Hydro sources')
plt.ylabel('Power Generation from Renewable sources')
plt.title('Correlation between Hydro and Renewable Power Generation')
plt.show()
```

```python
plt.plot(df['Month-Year'], df['Nuclear'],label='Nuclear plants')
plt.plot(df['Month-Year'],df['Renewable'],label='RE sources')
plt.plot(df['Month-Year'], df['Hydro'],label='Hydro Sources')
plt.xlabel('Month-Year')
plt.ylabel('Power Generation (BU)')
plt.title('Power Generation Hydro and Renewable sources over time')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
fig = px.line(df, x='Month-Year', y='Thermal')
fig = px.line(df, x='Month-Year', y='Total Power Generation(BU-Billion Units)')
fig.update_xaxes(type='category')
fig.update_layout(title='Total Power Generation over Time', xaxis_title='Month-Year', yaxis_title='Power Generation (BU)')
fig.show()
```

```python
sns.histplot(data=df, x='Thermal', kde=True)
plt.xlabel('Power Generation from Thermal sources')
plt.ylabel('Frequency')
plt.title('Distribution of Power Generation from Thermal sources')
plt.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=df['Month-Year'], y=df['Thermal'], name='Thermal', mode='lines+markers'))
fig.add_trace(go.Scatter(x=df['Month-Year'], y=df['Nuclear'], name='Nuclear', mode='lines+markers'))
fig.add_trace(go.Scatter(x=df['Month-Year'], y=df['Hydro'], name='Hydro', mode='lines+markers'))
fig.update_layout(title='Power Generation over Time', xaxis_title='Month-Year', yaxis_title='Power Generation (BU)')
fig.show()
```

```python


categories = ['Thermal', 'Nuclear', 'Hydro', 'Renewable']
values = df.iloc[0, 1:5].values

fig = go.Figure(data=go.Barpolar(
    r=values,
    theta=categories,
    marker_color='orange'
))
fig.update_layout(
    title='Polar Bar Plot',
    showlegend=False
)
fig.show()
```

```python
import plotly.express as px

fig = px.bar_polar(df, r='Total Power Generation(BU-Billion Units)', theta='Month-Year', color='Month-Year',
                   title='Bar Polar Plot', template='plotly_dark')

fig.show()
```

```python
import plotly.graph_objects as go

fig = go.Figure(go.Sunburst(
    labels=df['Month-Year'],
    parents=[''] * len(df),
    values=df['Total Power Generation(BU-Billion Units)'],
    branchvalues='total',
    marker=dict(colors=df['Total Power Generation(BU-Billion Units)'], colorscale='Blues')
))
fig.update_layout(
    title='Arc Plot for finding the top month with most production',
    height=600
)
fig.show()
```

```python
#modify it to get any category top months
fig = go.Figure(go.Icicle(
    labels=df['Month-Year'],
    parents=[''] * len(df),
    values=df['Thermal']
))
fig.update_layout(
    title='Icicle Plot for thermal electricity generation based on time of year',
    xaxis_title='Power Generation (BU)'
)
fig.show()
```

```python
fig = go.Figure()
for column in df.columns[1:]:
    fig.add_trace(go.Bar(x=df['Month-Year'], y=df[column], name=column))

fig.update_layout(title='Electricity Generation Over the sectors ',
                  xaxis_title='Year', yaxis_title='Value')
fig.show()
```
