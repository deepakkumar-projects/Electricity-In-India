```python
new_column_names = {
    'Net Electricity Generated from Utilities(GWH)': 'Net Electricity Generated from Utilities(GWH)',
    'Purchases from Non-Utilities + Net Import from Other Countries(GWH)': 'Purchases from Non-Utilities + Net Import from Other Countries(GWH)',
    'Net Electricity Available for Supply': 'Net Electricity Available for Supply',
    'Sold to Ultimate Consumers ': 'Sold to Ultimate Consumers',
    ' Loss in transmission & distribution': 'Loss in transmission & distribution',
    ' Loss in transmission & distribution     (%)': 'Loss in transmission & distribution(%)'
}

# Rename the columns in the DataFrame
df.rename(columns=new_column_names, inplace=True)
```

```python
plt.plot(df['Year'],df['Sold to Ultimate Consumers'],marker='^')
plt.plot(df['Year'], df['Net Electricity Generated from Utilities(GWH)'], marker='o',label='Utililiti')
plt.title('Net Electricity Generated from Utilities')
plt.xlabel('Year')
plt.ylabel('Net Electricity Generated (GWH)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
sns.barplot(data=df, x='Year', y='Net Electricity Generated from Utilities(GWH)')
plt.title('Net Electricity Generated from Utilities')
plt.xlabel('Year')
plt.ylabel('Net Electricity Generated (GWH)')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.set(style='whitegrid')

# Define the column names
columns = ['Net Electricity Generated from Utilities(GWH)', 'Sold to Ultimate Consumers']

# Create the stacked bar plot
sns.barplot(data=df, x='Year', y=columns[0], label=columns[0])
sns.barplot(data=df, x='Year', y=columns[1], bottom=df[columns[0]], label=columns[1])

# Add labels and title
plt.title('Net Electricity Generated and Sold to Ultimate Consumers')
plt.xlabel('Year')
plt.ylabel('Electricity Generated and Sold (GWH)')

# Show the legend
plt.legend()

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
sns.violinplot(data=df, x='Year', y='Sold to Ultimate Consumers')
sns.violinplot(data=df, x='Year', y='Net Electricity Generated from Utilities(GWH)')
plt.title('Net Electricity Generated from Utilities & Sold')
plt.xlabel('Year')
plt.ylabel('Net Electricity Generated (GWH)')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.barplot(data=df, x='Year', y='Loss in transmission & distribution')
plt.title('Loss in Transmission & Distribution')
plt.xlabel('Year')
plt.ylabel('Loss in Transmission & Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
sns.scatterplot(data=df, x='Year', y='Loss in transmission & distribution')
plt.title('Loss in Transmission & Distribution')
plt.xlabel('Year')
plt.ylabel('Loss in Transmission & Distribution')
plt.xticks(rotation=45)
plt.show()
```

```python
fig = px.line(df, x='Year', y='Loss in transmission & distribution', title='Loss in Transmission & Distribution')
fig.update_layout(xaxis_title='Year', yaxis_title='Loss in Transmission & Distribution')
fig.show()
```

```python
fig = go.Figure()
for column in df.columns[1:]:
    fig.add_trace(go.Scatter(x=df['Year'], y=df[column], mode='lines', name=column))

fig.update_layout(title='Electricity Generation and Consumption Over the Years',
                  xaxis_title='Year', yaxis_title='Value')
fig.show()
```

```python
fig = go.Figure()
for column in df.columns[1:]:
    fig.add_trace(go.Bar(x=df['Year'], y=df[column], name=column))

fig.update_layout(title='Electricity Generation and Consumption Over the Years',
                  xaxis_title='Year', yaxis_title='Value')
fig.show()
```

```python
fig = go.Figure()
for column in df.columns[1:]:
    fig.add_trace(go.Box(x=df['Year'], y=df[column], name=column))

fig.update_layout(title='Electricity Generation and Consumption Over the Years',
                  xaxis_title='Year', yaxis_title='Value')
fig.show()
```
