 ### Electric_energy_sales_India (2012-2013 to  2021-2022)
 ##### Columns - Description 
| Column Name     | Description    (GWH(Giga-Watts per Hour))                                              |
|-----------------|--------------------------------------------------------------|
| Domestic         | Ultimate Energy sold/consumed by Domestic Consumers (GWH)                                                |
| Commercial       | Energy Sold/Consumed By Commericial Entities (GWH)                  |
| Industrial(Low & Medium Voltage)     | Energy Sold/consumed by Low and Medium Voltage usage Industries |
| Industrial(High Voltage        |  Energy  Sold to High Voltage Usage Industries  |
| Public Lighting            | Public lighting Overall Sold Electricity(All_INDIA)       |
| Traction         | Energy Sold to transport based on Electricity(Trains,Vehichles)            |
| Agriculture       | Energy Sold/Consumed by Agricultue Activities  |
| Public Water Works & Sewage Pumping     | Consumed By Public Water Distribution & Seawage Pumping (Filters)  |
| Miscellaneous       | Energy Sold to all other Non-Primary Activities |
| Total Energy Sold (All_India IN GWH| Overall Sold/Consumed Electricity By Whole India           |

### Purpose 
* Analysis and prediction Of energy Sales for primary activities across INDIA
* EDA for Univariate/Bi-Variate Analysis
